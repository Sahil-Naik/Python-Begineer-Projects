```python
def srt(n1):
    if n1 > 0:
        r1 = n1 ** 0.5
        print("Square-root of given numbers is",r1)
    elif n1 == 0:
        print("Square-root of given number is 0")
    else:
        print("Square-root of Negative numbers is invalid")
        
def cbrt(n2):
    if n2 > 0:
        r2 = n2 ** (1/3)
        print("Cube-root of given numbers is",r2)
    elif n2 == 0:
        print("Cube-root of given number is 0")
    else:
        print("Cube-root of Negative numbers is invalid")

def sqr(n3):
    if n3 > 0:
        r3 = n3 ** 2
        print("Square of given number is :",r3)
    elif n3 == 0:
        print("Square of 0 is 0")
    else:
        print("Sqaure is invalid!")
        
def cbr(n4):
    if n4 > 0:
        r4 = n4 ** 3
        print("Cube of given number is :",r4)
    elif n3 == 0:
        print("Cube of 0 is 0")
    else:
        print("Cube is invalid!")

#Start of Program
print("\t _______Calculator_______")
a = int(input("Enter a number : "))
b = int(input("Enter a number : "))
res = 0
print("\n Choose a task to perform")
print("\n 1 = Addition \n 2 = Subtraction \n 3 = Division \n 4 = Square root \n 5 = Cube root \n 6 = Cube \n 7 = Sqaure")
#Choosing!!!!
c = int(input("Enter your choice : "))
if c == 1:
    print("You've choosen Addition!")
    res = a + b
    print("Addition of given numbers is",res)
elif c == 2:
    print("You've Choosen Subtraction!")
    res = a - b
    print("Subtracion of given number is",res)
elif c == 3:
    print("You've choosen Division!")
    res = a / b
    print("Division of given number is",res)
elif c == 4:
    print("You've choosen Square-root!")
    srt(a)
    srt(b)
elif c == 5:
    print("You've choosen Cube-root!")
    cbrt(a)
    cbrt(b)
elif c == 6:
    print("You've Choosen Cube!")
    cbr(a)
    cbr(b)
elif c == 7:
    print("You've Choosen Square!")
    sqr(a)
    sqr(b)
else: 
    print("Invalid Choice!")
```
