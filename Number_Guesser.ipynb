```python
import random

# python package 'random' is used  to generate a number between given intervals, for this program interval is 1 to 10
# i.e. random.randint(start, end)

generate = random.randint(0,10)
print("Guess the number generated by Computer between 1 to 10.")
guess = int(input("Enter your Guess:"))

if guess == generate:
    print("You've Guessed it correct! :)")
else:
    print("Computer generated {}, but you've guessed {}! :(".format(generate,guess))
```
